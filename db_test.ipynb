```python
import chromadb

persistent_client = chromadb.PersistentClient(path="vectorDB")
```

```python
collection_list = persistent_client.list_collections()
```

```python
name = collection_list[0].name
```

```python
col = persistent_client.get_collection(name = name)
```

```python
ergebnisse=[]
query_vek = embedding_function.embed_query(test_data[4])
for col_name in collection_list:
    name=col_name.name
    print(name)
    col = persistent_client.get_collection(name=name)
    res = col.query(query_vek,n_results=3)
    print(res)
    ergebnisse.append(res)
```

```python
len(ergebnisse)
```

```python
for e in ergebnisse:
    print(e)
```

```python
result = []
for e in ergebnisse:
    distances = e['distances'][0]
    
    indices = [idx for idx, value in enumerate(distances) if value < 1]
    if len(indices)>0:
        for idx in indices:
            item = { }
            
            item["ids"] = e['ids'][0][idx]
            item["distances"] = e['distances'][0][idx]
            item["metadatas"] = e['metadatas'][0][idx]
            item["documents"] = e['documents'][0][idx]
            
             
            result.append(item)

            print(item)
            print('****************')
    
```

```python


# Distance'a göre sıralama
sorted_list = sorted(result, key=lambda x: x["distances"])

print(sorted_list[:3])
```
